```python
import pandas as pd 

data = pd.read_csv("Dataset1.csv") 

# Preview the first 5 lines of the loaded data 
data.head()
```

```python
#drop useless columns
data = data.drop('education',axis=1)  #Equal to education-num
data = data.drop('fnlwgt',axis=1)     #Sampling Weight (not useful here)
data = data.drop('capital-gain',axis=1)
data = data.drop('capital-loss',axis=1)
data.head()
```

```python
#PREPROCESS
from sklearn.preprocessing import KBinsDiscretizer

def discritize(attr,n):
    c = data[attr].values.reshape(-1,1)
    kbin = KBinsDiscretizer(n_bins=n, encode='ordinal')
    kbin.fit(c)
    data[attr] = kbin.transform(c)
```

```python
#discritization of age
discritize('age',7)
data.head()
```

```python
#discritization of hours-per-week
discritize('hours-per-week',6)
data.head()
```

```python
y = data['income'].values #classification label
X = data.drop('income',axis=1).values
```

```python
#convert categorical/ordinal into one-hot
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
enc.fit(X)
X_oh = enc.transform(X).toarray()
```

```python
#use dicision tree to classify 
from sklearn import tree
#from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_oh, y, test_size=0.2, random_state=0)

clf = tree.DecisionTreeClassifier(random_state=0)
clf = clf.fit(X_train , y_train)
acc = clf.score(X_test , y_test)
print("Decision Tree Accuracy : ",acc)
```

```python
#plot graph
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
d = graph.render("dt_plot")
print("Plot saved in :",d)
```

```python
#use random forest to classify 
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X_oh, y, test_size=0.2, random_state=0)

clf = RandomForestClassifier(n_estimators=100,random_state=0)
clf = clf.fit(X_train , y_train)
acc = clf.score(X_test , y_test)
print("Random Forest Accuracy : ",acc)
```
